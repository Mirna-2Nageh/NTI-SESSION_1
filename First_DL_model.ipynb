```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense ,Flatten,Softmax
```

```python
#build a sequential model feedforward neural network model
model=Sequential([
Dense(16, activation='relu',input_shape=(28,28)),#inputlayer
Dense(16,activation='relu'),#hiddenlayer
Dense(10,activation='softmax')#outputlayer
])
```

```python
#print the model summary
print(model.weights)
```
